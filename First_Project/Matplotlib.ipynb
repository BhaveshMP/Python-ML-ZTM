## Using OO method

```python
hd.head(5)
```

```python
over_50 = hd[hd["age"] > 50]
over_50.head()
```

```python
over_50.plot(kind = 'scatter', x='age', y='chol', c='target')
```

```python
# OO method mixed with pyplot
fig, ax = plt.subplots(figsize = (10,5))
over_50.plot(kind = "scatter", x = "age", y="chol", c = "target", ax = ax)
ax.set_xlim([45,80])
```

```python
## OO method
fig, ax = plt.subplots(figsize = (10, 5))

scatter = ax.scatter(x = over_50["age"],
                    y = over_50["chol"],
                    c = over_50["target"])

ax.set(title = "Age vs Chol", xlabel = "Age", ylabel = "Chol")

#Adding legend
ax.legend(*scatter.legend_elements(), title = "Target" )

#Adding Horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle = '--')
```

```python
over_50.head(5)
```

```python
#subplots of chol, age, thalach

fig, (ax0, ax1)  = plt.subplots(figsize = (10, 10), ncols = 1, nrows = 2) 

ax0.scatter(x = over_50["age"], y = over_50["chol"], c = over_50["target"])
ax0.set(title = "Age vs Chol", xlabel = "Age", ylabel = "Chol")

ax0.legend(*scatter.legend_elements(), title = "Target" )

#Adding Horizontal line
ax0.axhline(over_50["chol"].mean(),
          linestyle = '--');

ax1.scatter(x = over_50["age"], y = over_50["thalach"], c = over_50["target"])

ax1.set(title = "Age vs Thalach", xlabel = "Age", ylabel = "Thalach")

ax1.legend(*scatter.legend_elements(), title = "Target")

ax1.axhline(over_50["thalach"].mean(), linestyle = "--")

#Adding tile to figure
fig.suptitle("Heart disease Analysis", fontsize = 18, fontweight = "bold")
```

```python
plt.style.use("seaborn-v0_8")
## OO method
fig, ax = plt.subplots(figsize = (10, 5))

scatter = ax.scatter(x = over_50["age"],
                    y = over_50["chol"],
                    c = over_50["target"],
                    cmap = "winter")

ax.set(title = "Age vs Chol", xlabel = "Age", ylabel = "Chol")

#Adding legend
ax.legend(*scatter.legend_elements(), title = "Target" )

#Adding Horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle = '--')
```

```python
#subplots of chol, age, thalach

fig, (ax0, ax1)  = plt.subplots(figsize = (10, 10), ncols = 1, nrows = 2) 

#Scatter plot 1
ax0.scatter(x = over_50["age"], y = over_50["chol"], c = over_50["target"], cmap = "winter")
ax0.set(title = "Age vs Chol", xlabel = "Age", ylabel = "Chol")

#Chaning x axis limit
ax0.set_xlim([50, 80])
#Setting Legend
ax0.legend(*scatter.legend_elements(), title = "Target" )

#Adding Horizontal line
ax0.axhline(over_50["chol"].mean(),
          linestyle = '--');



#Scatter plot 2
ax1.scatter(x = over_50["age"], y = over_50["thalach"], c = over_50["target"], cmap = "winter")

ax1.set(title = "Age vs Thalach", xlabel = "Age", ylabel = "Thalach")

#Chaning x axis limit
ax1.set_xlim([50, 80])
ax1.set_ylim([60, 200])

ax1.legend(*scatter.legend_elements(), title = "Target")

ax1.axhline(over_50["thalach"].mean(), linestyle = "--")

#Adding tile to figure
fig.suptitle("Heart disease Analysis", fontsize = 18, fontweight = "bold")

fig.savefig("Images/fig_23.jpg")
```
